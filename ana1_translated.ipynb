```python
import pandas as pd
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('data/comments_translated2.csv')
df.fillna('', inplace=True)
```

```python
chars = set()
for comment in df.translation:
    for c in comment:
        if not c in ['-']:
            if not c.isalpha():
                chars.add(c)
```

```python
my_stopwords = [
    'like', 'get','see','game','make',
    'know','video','videos','got','games',
    'would','one','oh','go','could',
    'right','ever','want','watch','people',
    'need','said','say','going','put','much',
    'best','better','playing', 'spam',
    'xd', 'watching','lol','really', 'made',
    'comment','--','even','thing','look',
    'omg','god','well','actually','something',
    'vc','way','favorite','next','keep',
    'good','use','ur','alot','ta','name','dont',
    'stuff','love','played','cool','play','im',
    'na', 'new', 'awesome','many', 'think', 'looks',
    'look','section', 'also', 'plz', 'great', 'nice',
    'give','upload','pls','channel','thanks','come',
    'haha','gg','take','yes','looking','bit','wan',
    'hey', 'bad', 'likes', 'already','lot','try',
    'tell', 'intro', 'saw','liked','end','nothing',
    'absolutely','definitely','tho','enough','btw',
    'though','thank','thanks','please','dd','exactly',
    'early','took','ya','etc','doesnt','dat','yo','mr',
    'came','thx','yay', 'gon'
]
```

```python
stop = \
    stopwords.words('english') + \
    my_stopwords
```

```python
def clean(s):
    r = s.lower().strip()
    for c in chars:
        r = r.replace(c, ' ')
    rs = [w for w in word_tokenize(r) if w not in stop and len(w) > 1]
    r = ' '.join(rs)
    return r
```

```python
%%time

df['cleaned'] = df.translation.apply(lambda x: clean(x) + ' ')
```

```python
df.to_csv('data/cleaned_translated_comments2.csv', index=False)
```
