```python
import networkx as nx
import shlex
import json
import csv
```

```python
fileName2='../minecraft_ch_data_clean.json'
fileName3='../overwatch_ch_data_clean.json'
g = nx.DiGraph()
```

```python
def addFile(fileName):
    with open(fileName) as f:
        for line in f:
            json_dict = json.loads(line)
            if (json_dict['statistics']['subscriberCount']!='0'):
                g.add_node(json_dict['id'], type='channel')
                g.node[json_dict['id']]['title']= json_dict['snippet']['title']
                g.node[json_dict['id']]['description']= json_dict['snippet']['description']
                g.node[json_dict['id']]['videoCount']= json_dict['statistics']['subscriberCount']
                g.node[json_dict['id']]['videoCount']= json_dict['statistics']['videoCount']
                g.node[json_dict['id']]['commentCount']= json_dict['statistics']['commentCount']
                g.node[json_dict['id']]['viewCount']= json_dict['statistics']['viewCount']
                # other attrs

                if 'keywords' in json_dict['brandingSettings']['channel']:
                    try:
                        for kw in shlex.split(json_dict['brandingSettings']['channel']['keywords']):
                            kw_id = kw.lower().replace("'", "").replace("´", "").replace("’", "")
                            g.add_node(kw_id, type='keyword')
                            g.add_edge(json_dict['id'], kw_id, type='has_kw')
                    except:
                        for kw in json_dict['brandingSettings']['channel']['keywords'].split():
                            kw_id = kw.lower().replace("'", "").replace("´", "").replace("’", "")
                            g.add_node(kw_id, type='keyword')
                            g.add_edge(json_dict['id'], kw_id, type='has_kw')

                if 'featuredChannelsUrls' in json_dict['brandingSettings']['channel']:
                    for ch in json_dict['brandingSettings']['channel']['featuredChannelsUrls']:
                        g.add_node(ch, type='channel')
                        g.add_edge(json_dict['id'], ch, type='features')
        
```

```python
addFile(fileName2)
g2=g.copy()   
```
