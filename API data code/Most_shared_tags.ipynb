```python
import csv
import pandas as pd

from pandas import Series
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.corpus import words
from collections import Counter
```

```python
filename1='../edges_corrected.csv'
filename2='../nodes3.csv'
df=pd.read_csv(filename1)
df2=pd.read_csv(filename2,index_col='Id')
community_num=[9,22,85,76,71,37,87]
```

## Get labels (edges) to detect community connections

```python
def set_modularity(identif):
    modularity=df2.at[identif, 'modularity_class']
    return modularity
```

```python
def clean(s):
    r = s.lower().strip()
    for c in chars:
        r = r.replace(c, ' ')
    rs = [w for w in word_tokenize(r) if w not in stop and len(w)>2]
    r = ' '.join(rs)
    return r+' '
```

```python
df['Label']= df.Source.apply(lambda x: set_modularity(x))
df['Id']= df.Target.apply(lambda x: set_modularity(x))
```

```python
chars=['{','}',',',':',"'"]
stop= set(stopwords.words('english')+['true'])
df['tags']=df['3'].apply(lambda x: clean(x))
```

```python
df3=df.groupby(['Label', 'Id']).tags.sum()
```

```python
t=['game','gaming','play','video','gameplay']
```

```python
edges=[(9,9),(22,22),(85,85),(76,76),(71,71),(37,37),(87,87)]
n_edges=[3902, 3159, 151, 1308, 315, 11566, 235]

m=0
for params, elem in df3.items():
    if params in edges:
        mshar=[]
        print(params)
        most_shared= Counter(Series(elem.split())).most_common()[0:20]  
        m_clean=list(most_shared)
        for tup in most_shared: 
            if tup[0] in t:
                m_clean.remove(tup)
                
        for tup in m_clean: 
            tag=tup[0]
            perc=round((tup[1]*100)/n_edges[m])
            perc=str(perc)+'%'
            
            mshar.append((tag,perc))
                
        for tup in mshar[0:20]:
            print(tup[0]+','+tup[1])
        m+=1
```

```python
com=[9,22,37,71,76,85,87]
n=[0,0,0,0,0,0,0]
for i in range(0, len(df)):
    for j in range (0, len (com)):
        if df.Label[i]==com[j] and df.Id[i]==com[j]:
            n[j]+=1
        
```
