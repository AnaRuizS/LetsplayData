```python
from nltk import word_tokenize
from nltk.corpus import stopwords
```

```python
original_df = pd.read_csv('data/comments_translated2.csv')
original_df.fillna('', inplace=True)
```

```python
chars = set()
for comment in original_df.translation:
    for c in comment:
        if not c in ['-']:
            if not c.isalpha():
                chars.add(c)
```

```python
def clean(s):
    r = s.lower().strip()
    for c in chars:
        r = r.replace(c, ' ')
    rs = [w for w in word_tokenize(r)] # if w not in stop and len(w) > 1]
    r = ' '.join(rs)
    return r
```

```python
%%time

original_df['cleaned'] = original_df.translation.apply(lambda x: clean(x) + ' ')
```

```python
%%time

original_comments_by_community_sr = original_df.groupby('community').cleaned.sum()
```

```python
%%time

for community, text in comments_by_community_sr.items():
    print('COMMUNITY:', community)
    counter = Counter(text.split())
    top_cleaned_50_words = [w for w, f in counter.most_common(50)]
    o_tokens = original_comments_by_community_sr[community].split()
    for token in top_cleaned_50_words:
        print('WORD:', token)
        counter = Counter([ng for ng in ngrams(o_tokens, 4) if token in ng])
        print(counter.most_common(10))
        print()
    print()
    print()
```
